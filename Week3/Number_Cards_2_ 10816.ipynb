```python
n = int(input())
n_list = list(map(int, input().split()))
m = int(input())
m_list = list(map(int, input().split()))

count_dic = {}

for i in n_list:
    if i in count_dic.keys():
        count_dic[i] += 1
    else:
        count_dic[i] = 1

for j in m_list:
    if j in count_dic.keys():
        print(count_dic[j], end = " ")
    else:
        print(0, end = " ")
```
