```python
# 틀린 답 - order_list를 reverse하지 않음.

num = int(input())
order_list = list(map(int, input().strip().split()))
order1_list = []
order2_list = []
order3_list = []

for i in range(num):
    if order_list[i] == 1:
        order1_list.append(num - i)
    elif order_list[i]== 2:
        order2_list.append(num - i)
    else:
        order3_list.append(num - i)

if order3_list != []:
    order3_list.reverse()

result_list = order1_list + order2_list + order3_list
print(*result_list)
```

```python
from collections import deque

num = int(input())
skills_list = list(map(int, input().strip().split()))
skills_list.reverse()
result_deque = deque()

for i in range(num):
    if skills_list[i] == 1:
        result_deque.appendleft(i + 1)
    elif skills_list[i] == 2:
        result_deque.insert(1, i + 1)
    else:
        result_deque.append(i + 1)

print(*result_deque)
```
