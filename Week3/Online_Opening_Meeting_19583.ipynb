```python
import sys

s, e, q = map(str, input().rstrip().split())
log_dic = {1:set(), 2:set()}  # set이 아니라 list로 할 경우 시간초과!

def time_calc(x):
    h, m = x.split(":")
    return int(h) * 60 + int(m)

for line in sys.stdin:
    if line == "\n":
        break

    chat_time, chat_name = line.split()

    if time_calc(chat_time) <= time_calc(s):
        log_dic[1].add(chat_name)  # set 자료형이므로 add로 추가

    elif time_calc(e) <= time_calc(chat_time) <= time_calc(q):
        if chat_name in log_dic[1]:
            log_dic[2].add(chat_name)


print(len(log_dic[2]))



```
