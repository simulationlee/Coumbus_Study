```python
n, m = map(int, input().rstrip().split())

set_dict = {}
count = 0

for i in range(n):
    input_str = input().strip()
    if input_str in set_dict.keys():
        set_dict[input_str] += 1
    else:
        set_dict[input_str] = 1

for j in range(m):
    input_str2 = input().strip()
    try:
        set_dict[input_str2] >= 1
        count += 1
    except:
        pass

print(count)
```
