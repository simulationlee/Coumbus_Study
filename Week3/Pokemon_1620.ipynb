```python
import sys

def input1():  # input 에러를 피하기위해 따로 정의
      return sys.stdin.readline()

N, M = map(int, input().split())
result_dic = {}
result_dic_reverse = {}

for i in range(1, N + 1):
    input_x = input1().strip()
    result_dic[i] = input_x
    result_dic_reverse[str(input_x)] = i  # value 값에서 key 값을 잧는거보다 2개의 dic를 정의함. (다른 방법은 없는가?)
for j in range(M):
    search = input1().strip()
    if search.isdigit():
        print(result_dic[int(search)])
    else:
        print(result_dic_reverse[search])

```
