```python
case_num = int(input())

for i in range(case_num):
    items_num = int(input())
    calc_dic = {}
    result = 1
    for j in range(items_num):
        k, v = map(str, input().strip().split())
        if v in calc_dic.keys():
            calc_dic[v].append(k)
        else:
            calc_dic[v] = []
            calc_dic[v].append(k)
    for key, value in calc_dic.items():
          result *= (len(value) + 1)
    result -= 1
    print(result)
```
